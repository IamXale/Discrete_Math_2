### **Task 1:**

Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

**Question 1: Show that the expected value** 

$$E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$$

To prove that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$, we will calculate $E[y]$ using the given joint probability distribution.

First, let's express $E[y]$ using the formula for expected value:

$$E[y] = \sum_{y} y \cdot P(y)$$

Where $y$ represents the vector of random variables $Y_1, Y_2, ..., Y_k$ such that $y = (y_1, y_2, ..., y_k)$.

Now, substitute the given joint probability distribution into the equation:

$$E[y] = \sum_{y} \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$$

Since $y_1 + y_2 + y_3 + ... + y_k = n$, we have:

$$E[y] = \sum_{y} \frac{n!}{y_1!y_2!y_3!...y_n!} (y_1 \cdot P_1)(y_2 \cdot P_2)...(y_k \cdot P_k)$$

Since the summation is over all possible combinations of $y_1, y_2, ..., y_k$ that satisfy $y_1 + y_2 + y_3 + ... + y_k = n$, each term in the summation represents a specific outcome of the experiment.

Now, observe that for each specific outcome, the product $(y_1 \cdot P_1)(y_2 \cdot P_2)...(y_k \cdot P_k)$ can be rearranged as $n \cdot p_1$ if the outcome corresponds to cell $i$ (where $i$ varies from 1 to $k$), because the $y_i$ and $P_i$ terms are independent of each other and $y_i$ equals the number of trials for which the outcome falls into cell $i$.

Therefore, the expected value $E[y]$ can be written as:

$$E[y] = \sum_{y} \frac{n!}{y_1!y_2!y_3!...y_n!} (n \cdot p_1) + \frac{n!}{y_1!y_2!y_3!...y_n!} (n \cdot p_2) + ... + \frac{n!}{y_1!y_2!y_3!...y_n!} (n \cdot p_k)$$

$$E[y] = n \cdot \left( p_1 + p_2 + ... + p_k \right)$$

$$E[y] = n \cdot 1$$

$$E[y] = n \cdot p_i$$

This concludes the proof that the expected value $E[y] = n \cdot p_i$.

**Question 1: Derive the probability distribution of the experiment.**

In this experiment, we have $n$ identical trials where each trial can result in either success ($S$) or failure ($F$) with probabilities $p$ and $q = (1 - p)$ respectively. Since the trials are independent, we can model the experiment using a binomial distribution.

The probability distribution of the experiment, denoted by $P(Y=y)$, where $Y$ is the number of successes observed during the $n$ trials, is given by the binomial probability mass function:

$$P(Y=y) = \binom{n}{y} \times p^y \times (1 - p)^{n-y}$$

where:
- $n$ is the number of trials,
- $y$ is the number of successes observed during the trials,
- $p$ is the probability of success on a single trial,
- $(1 - p)$ is the probability of failure on a single trial.

**Question 2: Show that the expectation of this probability distribution is $E[Y] = \Sigma y \times P(y) = np$.**

The expectation or mean of the binomial distribution is given by:

$$E[Y] = \sum_{y=0}^{n} y \times P(Y=y)$$

Substituting the expression for $P(Y=y)$, we get:

$$E[Y] = \sum_{y=0}^{n} y \times \binom{n}{y} \times p^y \times (1 - p)^{n-y}$$

To simplify this expression, we need to expand the binomial coefficient $\binom{n}{y}$, which represents the number of ways to choose $y$ successes out of $n$ trials:

$$
\binom{n}{y} = \frac{n!}{y!(n-y)!}
$$

Substituting this into the expression for $E[Y]$, we get:

$$
E[Y] = \sum_{y=0}^{n} y \times \frac{n!}{y!(n-y)!} \times p^y \times (1 - p)^{n-y}
$$

Now, let's simplify the terms:

$$
E[Y] = \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} \times p^y \times (1 - p)^{n-y}
$$

$$
= \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} \times p^y \times (1 - p)^{n-y} + 0 \times \frac{n!}{(0-1)!(n-0)!} \times p^0 \times (1 - p)^{n-0}
$$

$$
= \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} \times p^y \times (1 - p)^{n-y}
$$

Now, notice that when we multiply by $y$ and sum over $y=1$ to $n$, each term in the summation contains a factor of $y$:

$$
y \times \frac{n!}{(y-1)!(n-y)!} = n \times \frac{(n-1)!}{(y-1)!(n-y)!} = n \times \binom{n-1}{y-1}
$$

We can now rewrite $E[Y]$ as:

$$
E[Y] = n \times \sum_{y=1}^{n} \binom{n-1}{y-1} \times p^y \times (1 - p)^{n-y}
$$

This summation can be recognized as the expansion of $(p + (1 - p))^{n-1}$ using the binomial theorem. Since $p + (1 - p) = 1$, we have:

$$
E[Y] = n \times (p + (1 - p))^{n-1}
$$

$$
= n \times 1^{n-1}
$$

$$
= n
$$

So, we have shown that:

$$
E[Y] = n \times p

**Question 1: Find a general formula for the expected value of the distribution $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $.**

To find the expected value (mean) of a probability distribution, we use the formula:

$$ E(X) = \sum_{x} x \cdot P(X = x) $$

In this case, we're dealing with the Poisson distribution, where $ P(y) = \frac{\lambda^y}{y!} e^{-\lambda} $.

So, the expected value (mean) $ E(y) $ can be calculated as:

$$ E(y) = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $$

To simplify this expression, we can note that the sum is a series representation of the exponential function:

$$ e^{\lambda} = \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} $$

Now, differentiating both sides of this equation with respect to $ \lambda $, we get:

$$ \frac{d}{d\lambda} e^{\lambda} = \frac{d}{d\lambda} \left( \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} \right) $$
$$ e^{\lambda} = \sum_{y=1}^{\infty} \frac{y \cdot \lambda^{y-1}}{y!} $$

Multiplying both sides by $ e^{-\lambda} $, we get:

$$ e^{\lambda} \cdot e^{-\lambda} = \sum_{y=1}^{\infty} \frac{y \cdot \lambda^{y-1}}{y!} \cdot e^{-\lambda} $$
$$ 1 = \sum_{y=1}^{\infty} \frac{y \cdot \lambda^{y-1}}{y!} \cdot e^{-\lambda} $$

Now, multiplying both sides by $ \lambda $, we get:

$$ \lambda = \sum_{y=1}^{\infty} \frac{y \cdot \lambda^{y}}{y!} \cdot e^{-\lambda} $$

So, the expected value $ E(y) $ is simply $ \lambda $.

$$ E(y) = \lambda $$
